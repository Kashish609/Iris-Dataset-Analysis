```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/hp/Desktop/iris.csv", encoding='ascii', delimiter=',')
```

```python
#histogram of numerical data
```

```python
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for column in cols:
    plt.figure(figsize = (6,4))
    sns.histplot(data = data, x = column, kde = True, color = 'pink')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
numeric = data.select_dtypes(include=[np.number])
if numeric.shape[1] >= 4:
    plt.figure(figsize=(8, 6))
    corr = numeric.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap of Numeric Features')
    plt.tight_layout()
    plt.show()
```

```python
sns.pairplot(data, hue='species', height=2.5)
plt.suptitle('Pair Plot of Iris Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='species', y='sepal_length', data=data, palette='Set2')
plt.title('Box Plot of Sepal Length by Species')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.violinplot(x='species', y='petal_width', data=data, palette='pastel')
plt.title('Violin Plot of Petal Width by Species')
plt.tight_layout()
plt.show()
```
